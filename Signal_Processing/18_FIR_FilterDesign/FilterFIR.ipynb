```python
import scipy.signal as signal
import numpy as np
import matplotlib.pyplot as plt

from scipy.fft import fft, ifft
```

```python
w_size = 25
n_size = 1024

w = signal.get_window('hamming',w_size) # Ventana hamming

W = fft(w, n= n_size)
W_ =  (20*np.log((np.abs(W)))/2)[0:int(n_size/2)]

f = np.arange(0,1,1/int(n_size/2))

```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1,2,1)
plt.plot(w)
plt.subplot(1,2,2)
plt.plot(f,W_)
```

```python
norm_f = 0.5

fr = np.arange(0,1,1/25)

rect = np.zeros(w_size)
for i in range(len(rect)):
    if (fr[i] < norm_f):
        rect[i] = 1

RECT = ifft(rect, n= n_size)
# RECT_ =  (20*np.log((np.abs(RECT)))/2)[0:int(n_size/2)]


plt.figure(figsize=(10, 4))
plt.subplot(1,2,1)
plt.plot(rect)
plt.subplot(1,2,2)
plt.plot(RECT)
```

```python
h = w*RECT

H = fft(h, n= n_size)
H_ =  (20*np.log((np.abs(H)))/2)[0:int(n_size/2)]

fh = np.arange(0,1,1/int(n_size/2))
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1,2,1)
plt.plot(h)
plt.subplot(1,2,2)
plt.plot(fh,H_)
```
