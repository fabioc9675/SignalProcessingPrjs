# Solucion a ecuaciones en diferencias

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,10,0.1)
y = np.zeros(len(t))
Tm = 0.1

y[0]=1; y[1]=1-0.5*Tm
Tf = 10
c1 = 1+Tm+Tm*Tm
c2 = 2+Tm

for n in range(2,len(t)):
    y[n] = (y[n-1]*c2-y[n-2])/c1

plt.plot(t,y)
plt.grid(True)

```
