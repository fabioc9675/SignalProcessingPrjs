# ECG Signal generator 

This python script explore the generation of ECG signals to test different filtering algorithms

```python
import matplotlib.pyplot as plt
import numpy as np
from EcgGenerator import SimEcg
```

Generation of ECG D1

```python
# Datos para derivacion 1
a_D1 = np.array([.7, -1.0, 5.5, -1.0, 1.3, 0.1])
d_D1 = np.array([0.03, 0.011, 0.011, 0.011, 0.047, 0.0076])
t_D1 = np.array([0.11, 0.27, 0.3, 0.33, 0.605, 0.833])

ecgObj_D1 = SimEcg(a_D1, d_D1, t_D1, period=10)
ecg_D1, t = ecgObj_D1.ecg_gen()

plt.figure(figsize=(15, 2))
plt.plot(t, ecg_D1)
plt.title("ECG derivacion I")
```

Generation of ECG D2

```python
# Datos para derivacion 2
a_D2 = np.array([0.9, -2.2, 10, -2.2, 3.1, 0.2])
d_D2 = np.array([0.032, 0.013, 0.012, 0.013, 0.048, 0.0070])
t_D2 = np.array([0.10, 0.28, 0.32, 0.34, 0.62, 0.833])

ecgObj_D2 = SimEcg(a_D2, d_D2, t_D2, period=10)
ecg_D2, t = ecgObj_D2.ecg_gen()

plt.figure(figsize=(15, 2))
plt.plot(t, ecg_D2)
plt.title("ECG derivacion II")
```

Generation of ECG D3

```python
# Generation of derivations
ecg_D3 = ecg_D2-ecg_D1;

ecg_aVR = -(ecg_D1+ecg_D2)/2;
ecg_aVL = (ecg_D1-ecg_D2)/2;
ecg_aVF = (ecg_D2-ecg_D1)/2;
```

### Plot of the ECG Signals

```python
plt.figure(figsize=(15, 12))

plt.subplot(6,1,1)
plt.plot(t, ecg_D1)
plt.title("ECG derivacion I")

plt.subplot(6,1,2)
plt.plot(t, ecg_D2)
plt.title("ECG derivacion II")

plt.subplot(6,1,3)
plt.plot(t, ecg_D3)
plt.title("ECG derivacion III")

plt.subplot(6,1,4)
plt.plot(t, ecg_aVR)
plt.title("ECG derivacion aVR")

plt.subplot(6,1,5)
plt.plot(t, ecg_aVL)
plt.title("ECG derivacion aVL")

plt.subplot(6,1,6)
plt.plot(t, ecg_aVF)
plt.title("ECG derivacion aVF")
```

## Vectorcardiogram

```python
x_component = ecg_D1 + ecg_D2*np.cos(300*np.pi/180) + ecg_D3*np.cos(240*np.pi/180)
y_component = ecg_D2*np.sin(300*np.pi/180) + ecg_D3*np.sin(240*np.pi/180)

plt.figure(figsize=(6, 6))
plt.plot(x_component[0:250], y_component[0:250], linewidth=2 )

```
